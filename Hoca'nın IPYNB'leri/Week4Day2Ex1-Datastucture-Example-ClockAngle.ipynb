```python
def aciBul(saat,dakika,saniye):
    yelkovanAci = dakika * (360/60) + saniye * (6 /360)
    akrepEkstraAci = yelkovanAci * (30 / 360)
    akrepAci = saat * (360/12) 
    fark = yelkovanAci - (akrepAci + akrepEkstraAci)
    if fark < 0:
        fark = fark + 360
    return fark

aciBul(11,0,0)
```

```python
for saat in range(0,12):
    for dakika in range(0,60):
        for saniye in range(0,60):
            aci = aciBul(saat,dakika,saniye)
            if aci < 1:
                print(saat,":",dakika,":",saniye, "---> " ,aci)
```
