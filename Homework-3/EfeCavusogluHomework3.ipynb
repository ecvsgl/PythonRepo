```python
#Question1

def sameLetterCheck(word1, word2):
    set1=set()
    set2=set()
    charlist=list()
    for i in range(0,len(word1)):
        set1.add(word1[i])
    for i in range(0,len(word2)):
        set2.add(word2[i])
    if set1 != set2:
        return False
    charlist=list(set1)
    for i in range (0,len(charlist)):
        if word1.count(charlist[i]) == word2.count(charlist[i]):
            continue
        else:
            return False
    return True    
    
sameLetterCheck("kerem","ekrem")
```

```python
#Question2

def palindromes(filename):
    file = open(filename, "r")
    contents = file.readlines()
    palinset = set()
    for i in range(0, len(contents)):
        word = contents[i].strip("\n")
        palincheck = 1
        if len(word) %2 == 0:
            for j in range(0, int(len(word)/2)):
                if word[j].lower() != word[len(word)-1-j].lower():
                    palincheck = 0
            if palincheck == 1:
                palinset.add(word)
        if len(word) %2 != 0:
            for j in range(0, int((len(word)-1)/2)):
                if word[j].lower() != word[len(word)-1-j].lower():
                    palincheck = 0
            if palincheck == 1:
                palinset.add(word)
    
    return len(list(palinset))
    
palindromes("palindromes.txt")
```

```python
#Question 3

def distWords(filename):
    file = open(filename, "r")
    contents = file.readlines()
    deposet=set()
    for i in range(0,len(contents)):
        info = contents[i].split()
        for item in info:
            deposet.add(item)
    return deposet
            
    
distWords("meyveler.txt")
```

```python
#Question 4

def translator(fileinput, fileoutput):
    file1 = open(fileinput,"r")
    file2 = open(fileoutput, "w")
    contents = file1.readlines()
    translations={
        "uzum":"grape",
        "armut":"pear", 
        "nar":"pomegranade", 
        "ananas":"ananas", 
        "elma":"apple", 
        "karpuz":"watermelon", 
        "seftali": "peach", 
        "kayisi":"apricot", 
        "muz":"banana", 
        "portakal":"orange"}
    for i in range(0, len(contents)):
        translatedline=""
        words=contents[i].strip("\n").split(", ")
        print(words)
        for j in range(0,len(words)):
            words[j] = words[j].strip(",")
            if j==(len(words)-1) and i==(len(contents)-1):
                translatedline += translations[words[j]]
            elif j==(len(words)-1):
                translatedline += translations[words[j]] + ","
            else:
                translatedline += translations[words[j]] + ", "
        if i == len(contents)-1:
            file2.write(translatedline)
        else: file2.write(translatedline+"\n")
    file1.close()
    file2.close()
            
translator("turkish.txt", "english.txt")

a = open("english.txt", "r")
print(a.readlines())
```

```python
#Question 5

import numpy as np
import matplotlib.pyplot as plt

def dolarchart(filename):
    dataset = np.loadtxt(filename, delimiter="\t", dtype=('str','str'))
    datelist=[]
    fxlist=[]
    for i in range(0, len(dataset)):
        datelist.append(dataset[i][0])
        fxlist.append(float(dataset[i][1].replace(',','.')))   
    plt.figure(figsize=(20,20))
    plt.scatter(datelist,fxlist)
    plt.grid()
    plt.xticks([])
    plt.yticks([1,2,3,4,5,6,7,8,9],size=30,color='g')
    plt.show()

dolarchart("dolar.txt")
```
